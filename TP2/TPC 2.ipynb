```python
# Jogo Adivinha o número

import random

#  O utilizador tenta adivinhar o número que o computador pensou
def utilizador_adivinha_numero():
    x = random.randint(0, 100)  
    t = 0
    acertou = False

    print("Adivinha o número entre 0 a 100 que o computador pensou.")

    while not acertou:
        try:
            palpite = int(input("Qual é o seu palpite? "))
            t += 1

            if palpite == x:
                print(f"Acertou! O número era {x}. Você usou {t} tentativas.")
                acertou = True
            elif palpite < x:
                print("O número que pensei é Maior.")
            else:
                print("O número que pensei é Menor.")
        except ValueError:
            print("Por favor, insira um número válido.")

#   O computador tenta adivinhar o número que o utilizador pensou
def computador_adivinha_numero():
    minimo = 0
    maximo = 100
    tentativas = 0
    acertou = False

    print("Pense num número entre 0 e 100 e o computador vai tentar adivinhar!")

    while not acertou:
        palpite = (minimo + maximo) // 2 
        tentativas += 1
        print(f"O computador acha que o número é {palpite}.")

        resposta = input("Responda com 'Acertou', 'Maior' ou 'Menor': ").lower()

        if resposta == 'acertou':
            print(f"O computador acertou! O número era {palpite}. Ele precisou de {tentativas} tentativas.")
            acertou = True
        elif resposta == 'maior':
            minimo = palpite + 1
        elif resposta == 'menor':
            maximo = palpite - 1
        else:
            print("Resposta inválida. Por favor, responda com 'Acertou', 'Maior' ou 'Menor'.")

#  Escolher a modalidade
def jogar_adivinha():
    print("Bem-vindo ao jogo Adivinha o Número!")
    print("Escolha a modalidade:")
    print("1. O computador pensa num número, e você tenta adivinhar.")
    print("2. Você pensa num número, e o computador tenta adivinhar.")

    while True:
        try:
            escolha = int(input("Digite 1 ou 2 para selecionar a modalidade: "))
            if escolha == 1:
                jogador_adivinha_numero()
                break
            elif escolha == 2:
                computador_adivinha_numero()
                break
            else:
                print("Opção inválida. Por favor, escolha 1 ou 2.")
        except ValueError:
            print("Por favor, insira um número válido.")

# Iniciar o jogo
jogar_adivinha()
```
